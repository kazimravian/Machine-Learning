<b>Let's establish which variables in our Earthquake dataset are categorical and which are continuous. As we now know, categorical variables can also have numerical values, so having a numeric data type doesn't guarantee that a variable is continuous.</b>

1. Import the required libraries:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

2. Read the data. Use pandas' **.read_csv** method to read the CSV file into a pandas **DataFrame**:

```python
data = pd.read_csv("/content/drive/MyDrive/Machine Learning Lectures/datasets/earthquake_data.csv")
```

```python
# display first five rowss od dataset
data.head()
```

3. Find all the columns that are numerical and object types. We use the **.select_dtypes()** method on the DataFrame to create a subset DataFrame having numeric **(np.number)**. For numeric columns, use this command:

```python
numeric_variables = data.select_dtypes(include = [np.number])
numeric_variables.columns
```

4. For categorical columns, use categorical **(np.object)** columns, and then print the column names for each.

```python
object_variables = data.select_dtypes(include = [np.object_])
object_variables
```

5. Find the number of unique values for numeric features. We use the **unique** method on the DataFrame to find the number of unique values in each
column and sort the resulting series in ascending order. For numeric columns,
use this command:

```python
numeric_variables.nunique().sort_values()
```

6. For categorical columns, use this command:

```python
object_variables.nunique().sort_values()
```

<h1><font color = 'green' size = '6'>
<b>
Exercise 2: Calculating Category Value Counts
</b>
</font>
</h1>

For columns with categorical values, it would be useful to see what the unique values (categories) of the feature are, along with what the frequencies of these categories are, that is, how often does each distinct value occur in the dataset. Let's find the number of occurrences of each **0** to **4** label and **NaN** values for the **injuries_description** categorical variable.

```python
counts = data.injuries_description.value_counts(dropna = False)
counts
```

2. Sort the values in increasing order of the ordinal variable. If we want the
frequencies in the order of the values themselves, we can reset the index to give us a DataFrame and sort values by the index (that is, the ordinal variable):

```python
counts.reset_index().sort_values(by = 'index')
```

<h1><font color = 'green' size = '6'>
<b>
Exercise 3: Plotting a Pie Chart
</b>
</font>
</h1>

Let's plot a pie chart to visualize the distribution of the discrete values of the **damage_description**
variable.

1. Format the data into the form that needs to be plotted. Here, we run **value_
counts()** over the column and sort the series by index:

```python
counts = data.damage_description.value_counts()
counts.sort_index()
```

2. Plot the pie chart. The **plt.pie()** category plots the pie chart using the count data.

```python
fig, ax = plt.subplots(figsize = (5, 4))

slices = ax.pie(counts, labels = counts.index, colors = ['white'], wedgeprops = {'edgecolor' : 'black'})

patches = slices[0]
hatches = ['/', '\\', '|', '-']

for patch in range(len(patches)):
  patches[patch].set_hatch(hatches[patch])

plt.title('Pie chart showing counts for\ndamage_descriptioncategories')
plt.show()
```
