. In this exercise, we will look at a few important methods for performing advanced indexing and selecting data.

1. Create a list of the passengers' names and ages for those passengers under the age of 21, as shown here:

```python
child_passengers = df[df.Age < 21][['Name', 'Age']]
child_passengers
```

2. Count how many child passengers there were, as shown here:

```python
print(len(child_passengers))
```

3. Count how many passengers were between the ages of 21 and 30. Do this as follows:

```python
young_adult_passengers = df.loc[(df.Age > 21) & (df.Age < 30)]
len(young_adult_passengers)
```

4. Find the passengers who were either first- or third-class ticket holders. Do this as follows:

```python
df.loc[(df.Pclass == 1) | (df.Pclass == 3)]
```

5. Find the passengers who were not holders of either first- or third-class tickets.

```python
df.loc[~((df.Pclass == 1) | (df.Pclass == 3))]
```

6. We no longer need the **Unnamed: 0** column, so delete it using the **del** operator:

```python
del df['Unnamed: 0']
df.head()
```
