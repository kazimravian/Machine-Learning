We will study a handy method the **groupby** method, which provides a quick method for selecting groups of data at a time and provides additional functionality through the **DataFrameGroupBy** object.

1. Open a new Jupyter or colab notebook.

2. Use **pandas** to load the Titanic dataset:

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/Machine Learning Lectures/datasets/titanic.csv')
```

```python
df.head(5)
```

3. Use the **groupby** method to group the data under the **Embarked** column to find out how many different values for **Embarked** there are:

```python
embarked_grouped = df.groupby('Embarked')
print(f'There are {len(embarked_grouped)} Embarked group')
```

4. Display the output of **embarked_grouped.groups** to find what the **groupby** method actually does:

```python
embarked_grouped.groups
```

5. Use the **iloc** method to inspect row 1 and confirm that it belongs to embarked group **C**:

```python
df.iloc[1]
```

6. As the groups are a dictionary, we can iterate through them and execute
computations on the individual groups. Compute the mean age for each group,
as shown here:

```python
for name, group in embarked_grouped:
  print(name, group.Age.mean())
```

7. Another option is to use the **aggregate** method, or **agg** for short, and provide it with the function to apply across the columns. Use the **agg** method to determine the mean of each group:

```python
embarked_grouped.agg(np.mean)
```

8. Define a simple function that returns the first value in the column and then pass that function through to **agg**, as an example:

```python
def first_val(x):
  return x.values[0]

embarked_grouped.agg(first_val)
```

In this exercise, we will create a **Lambda** function that returns the first value in a column and use it with **agg**.

1. Write the **first_val** function as a Lambda function, passed to **agg**:

```python
embarked_grouped = df.groupby('Embarked')
embarked_grouped.agg(lambda x : x.values[0])
```

2. We can also pass multiple functions to **agg** via a list to apply the functions across the dataset. Pass the Lambda function as well as the NumPy mean and standard deviation functions, like this:

```python
embarked_grouped.agg([lambda x : x.values[0], np.mean, np.std])
```

3. Apply **numpy.sum** to the **Fare** column and the Lambda function to the **Age** column by passing **agg** a dictionary where the keys are the columns to apply the function to, and the values are the functions themselves to be able to apply different functions to different columns in the DataFrame:

```python
embarked_grouped.agg({
                      'Fare' : np.sum,
                      'Age' : lambda x : x.values[0]
                     })
```

4. Finally, execute the **groupby** method using more than one column. Provide the method with a list of the columns (**Sex** and **Embarked**) to **groupby**, like this:

```python
age_embarked_grouped = df.groupby(['Sex', 'Embarked'])
age_embarked_grouped.groups
```
